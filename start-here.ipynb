# Welcome to Sage Notebook MVP
1. What is Sage AI? Sage AI is powered by advanced reasoning modelsClaude 3.7 and Gemini 2.5designed specifically to understand data science, quant, and analytical problems. It helps you rapidly explore data, build models, and validate hypotheses with intelligent planning and execution.



2. How to Get Started:
* Open a new notebook file.
* Click on the **AI Chat** panel on the right.
* Clearly describe the problem or analysis youre working on.
* Provide context or instructions if your data has specific access requirements.
* The AI agent will always seek your confirmation before executing any cells. As this is an early-stage MVP, your review and active guidance are crucial.
3. Share Your Feedback: We value your insights! Please share what you find useful and what improvements or features youd like to see. Understanding your workflow greatly helps us enhance Sage.

## GOOD Prompting Techniques
1. Please be as specific explaining your goal as possible
2. The more detailed your plan and asks are the better it can plan
3. At each step you can **Reject** the current code and step taken by Sage-AI and prpose an improvement--this guidance is incredibly helpful for the best outcomes
## Not-So-Good Prompting
1. Vague asks without specific goals
2. Not asking to pull the right data
